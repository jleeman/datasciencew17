```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import sklearn as sk

path = '../data/titanic.csv'
titanic = pd.read_csv(path)

titanic.head()
```

```python
my_list = titanic['Survived'].value_counts() / titanic.shape[0]
print my_list
```

```python
X = titanic[['Pclass', 'Parch']]
y = titanic['Survived']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X, y)
```

```python
print logreg.coef_
```

```python
logreg.score(X_test, y_test)
```

```python
survived_pred_class = logreg.predict(X)
print survived_pred_class
```

```python
pclass_pred_prob = logreg.predict_proba(X_train)[:, 1]
print pclass_pred_prob
```
